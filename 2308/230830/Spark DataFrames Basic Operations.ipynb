# Spark DataFrames Basic Operations

```python
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder.appName('ops').getOrCreate()
```

```python
df = spark.read.csv('appl_stock.csv', inferSchema=True, header=True)
# inferSchema: json 파일에서는 사용할 수 없지만 csv파일에서는 사용가능
# 추가 설명은 없었지만 저번 수업에서 csv인 경우 스파크가 스키마추론을 쉽게 할 수 있다고 했다.
# header: csv파일의 첫 번째 행이 열의 이름이라는 것을 의미하는지 여부이다.
```

```python
df.printSchema()
# 파일은 주식에 대한 정보를 가지고 있다.
```

```python
df.show()
```

```python
df.head(3)[0]
```

```python
df.filter('Close < 500')
# DataFrame 반환
```

```python
# 데이터를 어떻게 필터링할지 의논한다.
# sql을 이미 알고 있다면 sql 명령을 통해 빠르고 쉽게 필터링할 수 있다.
df.filter('Close < 500').show()
```

```python
# 'Open' 컬럼만 
df.filter('Close < 500').select('Open').show()
```

```python
df.select('Open').filter('Close < 500').show()
# 예상대로 seleect와 filter의 순서를 바꾸어도 결과는 동일하다.
# select와 filter간 정해진 사용 순서는 없어보인다.
```

```python
# 'Close' 컬럼까지
df.filter('Close < 500').select(['Open', 'Close']).show()
```

```python
# sql이 아닌 파이썬 문법을 이용
df.filter(df['Close'] < 500).select('Volume').show()
```

```python
df.select('Volume').filter(df['Close'] < 500).show()
# 파이썬 문법을 사용할 때도 순서는 상관없다.
```

```python
# 틀린 예시
# df.filter(df['Close'] < 200 and df['Open'] > 200).show()
# and/or 대신 &/|를 사용해야한다.
```

```python
# 옳은 예시
df.filter((df['Close'] < 200) & (df['Open'] > 200)).show()
# 괄호로 조건을 확실히 구분해야한다.
# 부정은 조건앞에 ~을 붙인다.
```

```python
df.filter(df['Low'] == 197.16).show()
```

```python
df.filter(df['Low'] == 197.16).collect()
# show() 대신 collect()를 사용하면 실제 행 오브젝트 list가 반환된다.
# 즉, 결과를 변수에 저장할 수 있다.
# (그러면 show()는 변수에 저장할 수 없나보다.)
```

```python
result = df.filter(df['Low'] == 197.16).collect()
# 실전에서 대다수의 경우 추후에 변수로 작업할 수 있도록 collect를 사용할 것이다. 
# 이 코스에서는 show()를 사용하여 직접 결과를 볼 수 있도록 할 것이다.
```

```python
result
# 출력: 행 오브젝트가 값으로 들어간 리스트가 반환됨
```

```python
result[0]
# 출력: 행 오브젝트
```

```python
row = result[0]
```

```python
# 행 오브젝트를 dictionary로 변경
row.asDict()
# 매우 유용하다.
```
